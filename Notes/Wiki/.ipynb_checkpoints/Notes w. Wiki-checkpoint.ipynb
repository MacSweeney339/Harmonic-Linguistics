```python
import requests, time, datetime
from Wiki_Urls import Wiki_Urls
from pymongo import MongoClient
from bs4 import BeautifulSoup
```

```python
wiki = Wiki_Urls()
```

```python
wiki.artists_url
```

```python
artist_html = requests.get(wiki.artists_url).content
group_html = requests.get(wiki.groups_url).content
```

```python
artist_soup = BeautifulSoup(artist_html, 'html.parser')
group_soup = BeautifulSoup(group_html, 'html.parser')
```

```python
ass = artist_soup.select('.div-col.columns.column-width ul li a')
```

```python
gss = group_soup.select('.div-col.columns.column-count.column-count-3 ul li a')
```

```python
artist_list = [artist.text for artist in ass]
artist_list
```

```python
len(artist_list)
```

```python
group_list = [group.text for group in gss]
group_list
```

```python
len(group_list)
```

```python
len(artist_list) + len(group_list)
```
